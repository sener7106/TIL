```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
```

```python
data = datasets.load_breast_cancer()
```

```python
X = data.data
```

```python
y = data.target
```

```python
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
model = Sequential()
model.add(Dense(1, input_dim = 30, activation = 'sigmoid'))
```

```python
model.summary
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = Adam(learning_rate = 0.001), metrics = ['acc'])
```

```python
model.fit(X_train, y_train, epochs = 1000)
```

```python
X_test = scaler.transform(X_test)
```

```python
pred = model.predict(X_test)
```

```python
predict01 = np.where(pred > 0.5, 1, 0)
```

```python
predict01.flatten()
```

```python
np.sum((predict01.flatten() == y_test))
```

```python
import numpy as np
```

```python
X = np.array([[0., 0.],
              [0., 1.],
              [1., 0.],
              [1., 1.]])
```

```python
y = np.array([[0.],
             [1.],
             [1.],
             [0.]]) 
```

```python
y.shape
```

```python
w = np.array([[1], [2]], dtype = 'float32')
b = np.array([3], dtype = 'float32')
```

```python
w
```

```python
b
```

```python
np.dot(X,w) + b
```

```python
def sigmoid(h) :
  return 1 / (1 + np.exp(-h))
```

```python
hyp = np.dot(X, w) + b 
```

```python
#learning rate
learning_rate = 0.1

for i in range(1000) :
  for j in range(2) :
    hypothesis = sigmoid(np.dot(X, w) + b)
    print('예측값 : ', hypothesis)

    # 오차 계산
    cost = - 1 / 4 * (y * np.log(hypothesis) + (1 - y) * np.log(1 - hypothesis))
    print('cost : ', cost)
    # X의 줄, 칸을  변경하여 실행
    X_col_2d = X.T[j].reshape(1, -1)
    # gradient
    w_grad = 1 / 4 * np.sum(np.dot(X_col_2d, (hypothesis - y)))
    print('w_{}grad:{}'.format(i, w_grad))
    # update w
    w[j] = w[j] - learning_rate * w_grad
    print("w{} : {}".format(j, w))

    # gradient
    b_grad = 1 / 4 * np.sum(hypothesis - y)
    print('b_{}grad:{}'.format(i, b_grad))
    # update w
    b= b - learning_rate * b_grad
    print("b : {}".format(b))

```

```python
pred = sigmoid(np.dot(X, w) + b)
```
