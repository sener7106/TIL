# 귀무가설 vs 대립가설

__귀무가설__: 설정한 가설의 진실성 확률이 적어 처음부터 버릴 것으로 예상되는 가설 (H0)

__대립가설__: 가설검정에서 진위를 검증하고자하는 가설, 귀무가설이 기각될 때 받아들여지는 가설 (H1)

1. 강아지의 평균수명은 13년이라고 알려져 있다. 몇몇 전문가는 13년이 아니라고 주장한다. 따라서, 강아지의 평균수명이 실제로 어떻게 되는지를 검정하고자 한다. 이때 귀무가설과 대립가설을 설정하시오.

2. 전자제품의 불량률은 3%로 알려져 있다. 그런데 불량률에 대한 항의 민원 전화가 많이 와서, 불량률은 3%보다 더 클 수 있다는 소리가 나오고 있다. 이에 불량률이 실제로 어떻게 되는지 검정하고자 한다

3. 어느 공장에서 생산하는 제품의 분산은 7이라고 알려져 있다. 그런데 품질관리팀의 최근 조사에 의하면, 분산은 7보다 작을 수도 있다는 의견이 나왔다. 따라서, 제품의 분산이 실제로 어떠한지를 알아보기 위하여 검정을 시행하고자 한다.

H0: 강아지의 평균수명은 13년이다  
H1: 강아지의 평균수명은 13년이 아니다

H0 전자제품의 불량률은 3%보다 작거나 같다  
H1: 전자제품의 불량률은 3%보다 크다

H0: 제품의 분산은 7보다 크거나 같다  
H1: 제품의 분산은 7보다 작다

가설검정은 '같다','크다','작다'로 이루어집니다.  

```python
# 데이터 불러오기
df = pd.read_csv('C:/Users/USER/Desktop/강의진행/211208_솔트룩스_데이터분석강의/data/data_norm.csv')

# 정규성 검정
x = stats.shapiro(df)
print('The Shapiro-Wilkis Statistic is %.3f and the p-value is %.3f' %x)

# 1-Sample t-test
t_result = stats.ttest_1samp(df, 76.7)
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)
print( "1-Sample t-test ")
print("t검정통계량 : {}".format(t))
print("p-value : {}".format(p))
```

```python
# 데이터 불러오기 
df1 = pd.read_csv('C:/Users/USER/Desktop/강의진행/211208_솔트룩스_데이터분석강의/data/data_sampt1.csv')
df2 = pd.read_csv('C:/Users/USER/Desktop/강의진행/211208_솔트룩스_데이터분석강의/data/data_sampt2.csv')

# 정규성 검정
x=stats.shapiro(df1)
print('Shapiro-Wilkis검정통계량은 %.3f, p-value는 %.3f' %x)


y=stats.shapiro(df2)
print('Shapiro-Wilkis검정통계량은 %.3f, p-value는 %.3f' %y)


# 등분산성 검정
stat, p = stats.bartlett(df1['sample'],df2['sample'])
print('등분산검정결과 p-value는 %.3f' % p)

# 2-Sample t-test
t_result = stats.ttest_ind(df1['sample'], df2['sample'], equal_var = True)
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)

print( "2-Sample t-test ")
print( "t검정통계량: {} " .format(t))
print( "p-value: {} " .format(p))
```

[정규성 검정]
- 귀무가설(H0): 정규성이 있다. 
- 대립가설(H1): 정규성이 없다.

[등분산성 검정]
- 귀무가설(H0): 등분산성이 있다.(분산의 차이가 없다) 
- 대립가설(H1): 등분산성이 없다.(분산의 차이가 있다)

__결과 해석__
- 집단 A는 정규성을 만족한다.(유의수준 0.05)
- 집단 B는 정규성을 만족하지 않는다. (유의수준 0.05)
- 집단 A와 B는 등분산성을 만족한다.(분산의 차이가 없다)

__고객 만족도 설문조사 A고객과 B고객 모평균 차이가 있는가?__  
H0: A고객과 B고객은 모평균이 같다.  
H1: A고객과 B고객은 모평균이 다르다.

=> P-value가 0.003으로 귀무가설 기각(대립가설 체택)  
=> A고객과 B고객의 모평균은 차이가 존재한다.

```python
# 데이터 불러오기 
df = pd.read_csv('C:/Users/USER/Desktop/강의진행/211208_솔트룩스_데이터분석강의/data/data_pair.csv')

# paired t-test
t_result = stats.ttest_rel(df['before'], df['after'])

t, p = t_result.statistic.round(3), t_result.pvalue.round(3)

print( "Paired t-test ")
print( " t검정통계량 : {} " .format(t))
print( " p-value : {} " .format(p))
```

__p-value = 0.0 < 0.05이므로 유의수준 0.05에서 귀무가설을 기각한다.__
