```python
#Ex 3-2. 물품 대분류를 기준으로 구매금액에 대한 합 출력
pd.pivot_table(data = table1, index = '물품대분류', values = '구매금액', aggfunc = 'sum')
```

```python
#Ex3-3. 모든 데이터에 대한 요약 통계량 확인하기

table1.describe(include = 'object')
```

```python
#Ex 3-4. 예제 데이터의 성별을 1과 0으로 라벨링하기
table1['성별(재정의)'] = table1['성별'].replace('남', 1).replace('여', 0)
table1['성별'].value_counts(normalize = True)
```

```python
#Ex 3-5. 구매 수량이 음수(-)인 값들만 추출하여 테이블 생성
table2 = table1[table1['구매수량'] < 0]
```

```python
print(table2.shape)
table2.head()
```

```python
# Ex 3-6. 예제 데이터에서 회원번호를 기준으로 구매금액의 합을 계산하고 가장 많이 구매한 고객 5명을 확인
table3 = pd.pivot_table(data = table1, index = '회원번호', values = '구매금액', aggfunc = 'sum')
```

```python
table3.sort_values(by = '구매금액', ascending = False).head(5)
```

```python
# Ex 3-7. '요일'별로 '물품대분류'의'구매금액'의 합을 계싼하고 판매량이 가장 높은 요일과, 해당 요일에 가장 많이 판대묀 품목 확인
table5 = pd.pivot_table(data = table1, index = '요일', columns = '물품대분류', values = '구매금액', aggfunc = ['sum','max']).T
table5
```
