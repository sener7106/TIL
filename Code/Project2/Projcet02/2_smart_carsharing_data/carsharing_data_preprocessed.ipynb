# 2. 데이터 병합
1. concat: 데이터 프레임 두 개를 단순히 합칠 때 사용한다.
2. merge: concat이 단순히 두 데이터 프레임을 합치는 거라면, merge는 두 데이터 프레임을 공통된 항목을 기준으로 합치는 것이다.
 - inner: 두 데이터 프레임의 교집합을 출력
 - left: 두 데이터 프레임의 교집합뿐만 아니라, 왼쪽 데이터 프레임에는 있고 오른쪽 데이터 프레임에는 없는 정보도 모두 출력한다.

```python
result01 = pd.merge(df, df_, on = '아이디', how = 'left')
print(result01.shape)
result01.head()
```

```python
result01.columns
```

__[중복 데이터 삭제]__  
- rename columns: '성별_x','거주단지_x','가입년월_x'  
- drop columns: '가입년월_y','성별_y','거주단지_y'

```python
result02 = result01.drop(columns = ['가입년월_y','성별_y','거주단지_y'])
result03 = result02.rename({'성별_x':'성별', '거주단지_x':'거주단지', '가입년월_x':'가입년월'})
print(result03.shape)
```
