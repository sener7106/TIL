## 데이터선택

- filter : 이름을 이용한 컬럼을 선택, 데이터의 특정 컬럼을 선택하거나 특정 문자를 
    포함하고 있는 컬럼 선택 및 데이터 추출
- argument .. items : 직접 입력 컬럼 선택, like = 입력 값을 포함한 컬럼 선택
- mask : 조건에 해당하는 데이터 행(row)을 mask 처리 .. cond = 마스크 적용 조건, other = 적용 조건 시 처리


```python
df = pd.read_csv(f'{path}/data_fil.csv')
```

```python
df.shape
df
```

```python
df.filter(items = ['ABC', 'value1'])
df.filter(regex = '1', axis = 1)
# regex 이용. '1'로 끝나는칼럼
df.filter(regex = '1$', axis = 1)
```

```python
df.filter(like ='a', axis = 1)
df.filter(regex = '[Aa]', axis =1)
```

```python
df.mask(df.value1 > 12, np.NaN)
```
