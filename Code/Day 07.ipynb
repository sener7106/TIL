# 귀무가설 VS 대립가설 : 추론통계..Estimation

<b>귀무가설 : </b> 설정한 가설의 진실성 확률이 적어 처음부터 버릴 것으로 예상되는 가설, 검정의 대상으로 삼는 가설

__대립가설__: 가설검정에서 진위를 검증하고자 하는 가설, 귀무가설이 기각 될때 받아들여지는 가설

 1. 강아지의 평균 수명은 13년이라고 알려져 있다. 몇 전문가는 13년이 아니라고 주장한다. 따라서 강아지의 평균수명은 실제로 어떻게 되는지를 검정하고 싶다.
 
 - 강아지의 평균 수명은 13년이다. -> 귀무가설
 - 강아지의 평균 수명은 13년이 아니다 -> 대립가설
 
 2. 전자제품의 불량률은 3%로 알려져 있다. 그런데 불량률에 대한 항의 민원 전화가 많아 불량이 3% 보다 더 클 수 있다는 소리가 나오고 있다. 이에 불량률이 실제로 어떻게 되는지 검정하고자 한다.
  - 제품 불량률은 3%보다 작거나 같다 -> 귀무가설
  - 3% 보다 크다-> 대립가설
  
 3. 어느 공장서 생산하는 제품의 분산은 7 이라고 알려져 있다. 그런데 품질관리팀의 최근 조사에 의하면, 분산은 7보다 작을 수도 있다는 의견이 나왔다. 따라서. 제품의 분산이실제로 어떠한지를 알아보기 위하여 검정을 시행하고자 한다.
 
  - 제품의 분산은 7보다 크거나 같다. -> 귀무가설
  - 제품의 분산은 7보다 작다. -> 대립가설
  
 ### 귀무가설(Null Hypothesis : 이론에 대한 가설), 대립 가설(Alternative Hypothesis : 이론에 대한 반론)
 
  __P-value__ : 귀무가설이 참이라는 가정하에 표본데이터가 귀무가설을 지지하는 확률, 이 값이 유의수준(알파)보다 작으면 귀무가설을 기각함(대립 가설을 지지).

```python
df = pd.read_csv(f'{path}/data_norm.csv')
# 정규성 검정 - 정규분포에 따르는가?
# 고객만족도 평균은 76.7이 맞다(귀무가설)
x = stats.shapiro(df) # shapiro-wilk test
print('The Shapiro-wilkis Statistic is %.3f and the p-value is %.3f' % x)
```

```python
t_result = stats.ttest_1samp(df, 76.7)
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)
print("1-Sample t-test")
print("t검정통계량 : {}".format(t))
print("p-value : {} ".format(p))
```

- alpha = 0.05, 보다 P-value(0.41, 0.480) 으로 크기 때문에 귀무가설은 적절하다. 즉, 평균은 76.7이다 -> 만족도 변화 x

```python
# 고객만족도 확인을 위해 A와 B고객에 대해 조사하여 다음 데이터를 얻음.
# A고객과 B고객의 모평균에 차이가 있는가?  - 귀무가설. 모평균에 차이는 없음
```

```python
df1 = pd.read_csv(f'{path}/data_sampt1.csv')
df2 = pd.read_csv(f'{path}/data_sampt2.csv')

# 정규성 검정
x = stats.shapiro(df1)
print('Shaprirpo-Willkis검정 통계량은 %.3f, p-value는 %.3f' %x)
y = stats.shapiro(df2)
print('Shaprirpo-Willkis검정 통계량은 %.3f, p-value는 %.3f' %y)

stat, p = stats.bartlett(df1['sample'], df2['sample'])
print('등분산 검정결과 p -value 는 %.3f' %p)

# 0.05 < 0.642 , 알파보다 크므로, 정규
t_result = stats.ttest_ind(df1['sample'], df2['sample'], equal_var = True)
t, p = t_result.statistic.round(3), t_result.pvalue.round(3)
print( "2-Sample t-test ")
print( "t검정통계량: {} " .format(t))
print( "p-value: {} " .format(p))

# p-value 가 알파 보다 작으므로, 모평균에 차이가 있다.

df1.describe()
```
