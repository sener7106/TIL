```python
import numpy as np
import tensorflow as tf
```

```python
tf.enable_eager_execution()
print(tf.__version__)
```

```python
# AND 연산의 x1, x2의 입력을 저장

X = np.array([[0.,  0.],
              [0., 1.],
              [1., 0.],
               [1., 1.]], dtype = 'float32')
```

```python
y = np.array([
              [0.],
              [0.],
              [0.],
              [1.]
], dtype = 'float32')
```

```python
w = tf.Variable([[1],[2]], dtype = 'float32')
b = tf.Variable([3], dtype = "float32")
```

```python
# X와 w의 곱을 계산
tf.matmul(X, w) 
```

```python
# tf.matmul (X, w) + b
hypothesis = tf.sigmoid(tf.matmul(X, w) + b)
```

```python
# W와 b값을 자동으로 수정
optimizer = tf.train.AdamOptimizer(0.01)
```

```python
for step in range(5000) :
  with tf.GradientTape() as tape :
    hypothesis = tf.sigmoid(tf.matmul(X, w) + b)
    cost = -tf.reduce_mean(y * tf.log(hypothesis) + (1 - y) * tf.log(1 - hypothesis))
    # cost W, b를 이용하여 W의 기울기, b의 기울기를 계산해서 grad에 대입
    grads = tape.gradient(cost, [w,b])
    optimizer.apply_gradients(grads_and_vars = zip(grads, [w, b]))


  if step % 100 == 0 :
    print("=" * 40)
    print("step : {}, cost: {}, w : {}, b{}".format(step, cost, w.numpy(), b.numpy()))
    print("=" * 40)
```

```python
predict = tf.sigmoid(tf.matmul(X, w) + b)
predict
```

```python
predict01 = tf.cast(predict > 0.5 , dtype = tf.float32)
print("===================")
print("predict01")
print(predict01)
```

```python
# tf.equal(predict, y_data) : 조건을 만족하면 true or False 리턴

ac01 = tf.equal(predict01, y)
ac01
```

```python
# tf.cast() = bool -> float

ac02 = tf.cast(ac01, dtype = 'float32')
ac02
```

```python
ac03 = tf.reduce_mean(ac02)
ac03
```

```python
accuracy = tf.reduce_mean(tf.cast(tf.equal(predict01, y), dtype = tf.float32))
print("=======")
print("accuracy")
print(accuracy.numpy())
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
tf.enable_eager_execution()
```

```python
data = datasets.load_breast_cancer()
```

```python
data
```

```python
X = data.data
X
```

```python
X = np.array(X, dtype = 'float32')

X
```

```python
y = data.target
```

```python
y = np.array(y, dtype = 'float32')
```

```python
y = y.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
def sigmoid(h ) :
  return 1 / (1 + np.exp(-h))
```

```python
#tf.contrib.layers.xavier_initializer()
initializer = tf.contrib.layers.xavier_initializer()

w = tf.Variable(initializer([30, 1]))
w
```

```python
b = tf.Variable(initializer([1]))
b
```

```python
hypothesis = tf.matmul(X_train, w) + b
hypothesis
```

```python
optimizer = tf.train.AdamOptimizer(0.001)
```

```python
for step in range(100) :
  with tf.GradientTape() as tape :
    hypothesis = tf.sigmoid(tf.matmul(X_train , w) + b)

    cost = -tf.reduce_mean(y_train * tf.log(hypothesis) + (1 - y_train) * tf.log(1 - hypothesis)) 

    grads = tape.gradient(cost, [w, b])

  optimizer.apply_gradients(grads_and_vars = zip(grads, [w,b]))

  if step % 10 == 0 :
    print("=" * 50)
    print("step : {}, cost : {}, w {}, b{}".format(step, cost.numpy(), w.numpy(), b.numpy()))
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
pred = tf.sigmoid(tf.matmul(X_test, w) + b)
pred
```

```python
pred01 = tf.cast(pred > 0.5, dtype = tf.float32)
print("pred01")
print(pred01)
```

```python
ac01 = tf.equal(pred01, y_test)
```

```python
ac02 = tf.reduce_mean(tf.cast(ac01, dtype = "float32"))
ac02
```

```python
accuracy = ac02
print('accuracy : ', accuracy.numpy())
```
